```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

numerical_cols = ['Price', 'living area', 'lot area', 'number of bedrooms', 
                  'number of bathrooms', 'grade of the house']

print("Outlier Analysis:")
print("=" * 50)
for col in numerical_cols:
    outliers, lower, upper = detect_outliers_iqr(data, col)
    print(f"{col}:")
    print(f"  Total outliers: {len(outliers)} ({len(outliers)/len(data)*100:.2f}%)")
    print(f"  Range: [{lower:.2f}, {upper:.2f}]")
    print()
```

```python

def cap_outliers(data, column, factor=1.5):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - factor * IQR
    upper_bound = Q3 + factor * IQR
    
    data[column] = data[column].clip(lower=lower_bound, upper=upper_bound)
    return data

data_processed = data.copy()

outlier_cols = ['living area', 'lot area', 'number of bedrooms', 'number of bathrooms']
for col in outlier_cols:
    data_processed = cap_outliers(data_processed, col)
```

```python
data_processed['price_per_sqft'] = data_processed['Price'] / data_processed['living area']
data_processed['house_age'] = 2024 - data_processed['Built Year']  
data_processed['renovation_age'] = 2024 - data_processed['Renovation Year']
data_processed['renovation_age'] = data_processed['renovation_age'].clip(lower=0)
data_processed['total_area'] = data_processed['living area'] + data_processed['Area of the basement']
```

```python

selected_features = [
    'living area', 'number of bedrooms', 'number of bathrooms',
    'grade of the house', 'condition of the house', 'Number of schools nearby',
    'Distance from the airport', 'price_per_sqft']
```

```python
x=data_processed[selected_features]
y=data[['Price']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled= scaler.transform(X_test)
print(f"\nmean of each feature: {scaler.mean_}")
print(f"\nstandard deviation of each feature: {scaler.scale_}")
```

## baseline model: Ridge Regression

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso

ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)
y_pred_ridge = ridge.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
rmse_ridge = mean_squared_error(y_test,y_pred_ridge)
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
print(f"RMSE: {rmse_ridge:.2f}")
print(f"MAE: {mae_ridge:.2f}")
print(f"R² Score: {r2_ridge:.2f}")
accuracy_percentage = r2_ridge * 100
print(f"Model Accuracy: {accuracy_percentage:.2f}%")
```

## Training other Advanced models

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
models = {
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=200, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=200, learning_rate=0.1, max_depth=8, random_state=42, n_jobs=-1),
    }
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} → RMSE: {rmse:.2f}, R²: {r2:.2f}")
```

# Hyperparameter Tuning


```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('ridge', Ridge())
])
param_grid = {'ridge__alpha': [0.1, 1.0, 10.0, 100.0]}
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
best_ridge_model = grid_search.best_estimator_
y_pred = best_ridge_model.predict(X_test)
print(f"R-squared: {r2_score(y_test, y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
final_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('ridge', Ridge(alpha=grid_search.best_params_['ridge__alpha']))
])
final_pipeline.fit(x, y)
```

```python
import joblib
from datetime import datetime

# Save your best Ridge model (from GridSearchCV)
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
model_filename = f"best_ridge_{timestamp}.joblib"

joblib.dump(best_ridge_model, model_filename)
print(f"✅ Model saved as {model_filename}")
```
